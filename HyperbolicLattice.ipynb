```python
def TOTAL_ENERGY(graph, spin_map):
    #the total energy is the sum of all spins on the map
    energy = 0
    
    #iterate each graph element and get its neightbors
    #compute the energy and then continue to the next one
    _SIZE = len(graph)
    for i in range(_SIZE):
        #get the neightbors
        si = spin_map[i] #curent spin
        e = 0
        for n in graph[i]: #for each neighbors 
            sj = spin_map[n]
            e = e + si * sj
        
        
        
        energy = energy + e
        
    return energy/2
```

```python
spins = np.random.choice(a=[False, True], size=_size)
TOTAL_ENERGY(graph, spins)
```
